# Box Plot

```python
import  matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70,130]

plt.boxplot(x,notch=True,vert=False,widths=0.5,labels=['Python'],
            patch_artist=True,showmeans=True,sym='go',
            boxprops={'color':'r'},capprops={'color':'g'},
            whiskerprops={"color":'y'})
plt.show()
```

**Multiple Box Plots**

```python
x=[10,20,30,40,50,60,70,130]
x1=[30,20,40,50,70,10,80,90]

y=[x,x1]

plt.boxplot(y,notch=True,vert=True,widths=0.5,labels=['Python','C++'],
            patch_artist=True,showmeans=True,sym='go',
            boxprops={'color':'r'},capprops={'color':'g'},
            whiskerprops={"color":'y'})



plt.show()
```
