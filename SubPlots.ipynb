# SUBPLOTS

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[1,2,3,4]

plt.plot(x,y,color='r') # Line Plot

plt.pie([1],colors='b') # Dot chart

plt.pie(x,radius=0.5) # Pie Chart

plt.bar(x,y) # Bar Char

'''Here all the charts are getting overlapped, but i want them to be separated.
For this purpose we will use Subplots.'''

plt.show()
```

```python
# plt.subplots(nrows,ncols,index)

x=[1,2,3,4]
y=[1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x,y,color='r')
plt.savefig('Line Chart')


plt.subplot(2,2,2)
plt.pie([1])
plt.savefig('Dot Pie Chart')

plt.subplot(2,2,3)
plt.pie(x)
plt.savefig('Pie Chart')

plt.subplot(2,2,4)
plt.bar(x,y)
plt.savefig('Bar Chart')

plt.savefig('Subplots Figure',dpi=1200,facecolor='g')
plt.savefig('Subplots Figure.pdf',dpi=1200,transparent=True)


plt.show()

'''Here in single figure, we have multiple subplots.'''
```
