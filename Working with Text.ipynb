# Working with Text in Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,5,1,4,2]

plt.plot(x,y,marker='o')

plt.title("Growth of Python",fontsize=15,color='g')
plt.xlabel('Days')
plt.ylabel('Python')

plt.text(2,5,'Python',fontsize=12,style='italic',bbox={'facecolor':'cyan'})

plt.annotate('C++',xy=(1,3),xytext=(2,1),arrowprops={'facecolor':'black'})

plt.legend('growth',loc=10,shadow=True)

plt.show()
```
