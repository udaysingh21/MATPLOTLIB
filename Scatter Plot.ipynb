# SCATTER PLOT

```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7]
num=[2,8,1,7,3,9,6]

plt.scatter(day,num,color='r')

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.show()
```

```python
day=[1,2,3,4,5,6,7]
num=[2,8,1,7,3,9,6]

colors=[10,49,30,69,56,20,30]
size=[200,800,100,700,300,900,600]

# plt.scatter(day,num,c=colors,s=size,edgecolor='g',linewidths=2)
plt.scatter(day,num,c=colors,s=size,cmap='Accent_r')
t=plt.colorbar()
t.set_label("ColorBar")

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.show()
```

**Multiple data in single scatter plot**

```python
day=[1,2,3,4,5,6,7]
num=[2,8,1,7,3,9,6]
num1=[3,5,2,1,8,6,4]

colors=[10,49,30,69,56,20,30]
size=[200,800,100,700,300,900,600]

# plt.scatter(day,num,c=colors,s=size,edgecolor='g',linewidths=2)
plt.scatter(day,num,c=colors,s=size,cmap='Accent_r')
plt.scatter(day,num1,c='r',s=size)
t=plt.colorbar()
t.set_label("ColorBar")

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.show()
```
