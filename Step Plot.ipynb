# Step Plot  
*It is useful when you want to see changes that occur at regular intervals*  
*For ex: price rise in milk products, tax rates, interest rates, etc.*

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,2,5,4,1]

# ms- marker size
# mfc - marker face color
plt.step(x,y,color='g',marker='o',ms=10,mfc='r',label='Python')

plt.title("Python")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc=1)

plt.grid()
plt.show()
```
