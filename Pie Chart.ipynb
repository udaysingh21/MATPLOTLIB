# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
labels=['c','c++','java','python']
explode=[0,0,0,0.08]
# colors=['r','g','b','y']

plt.pie(x, labels=labels, explode=explode,autopct='%0.1f%%', shadow=True, radius=1.1,
        labeldistance=1.1, startangle=90, textprops={'fontsize':12}, counterclock=False,
        wedgeprops={'linewidth':3},rotatelabels=False)

plt.title("Pie Chart")
plt.legend(loc=1)
plt.show()
```

**Dot Pie Chart**

```python
plt.pie([1])
plt.show()
```

**Donut Pie Chart**

```python
x=[10,20,30,40]
x1=[20,40,25,15]
labels=['c','c++','java','python']
explode=[0,0,0,0.08]
colors=['r','c','b','y']

plt.pie(x, labels=labels, radius=1)
plt.pie(x1,radius=0.5,colors=colors,labels=labels)

plt.title("Pie Chart")
plt.show()
```

**Ring Pie Chart**

```python
x=[10,20,30,40]

labels=['c','c++','java','python']
explode=[0,0,0,0.08]
colors=['r','c','b','y']

plt.pie(x, labels=labels, radius=1.2)
plt.pie([1],radius=0.8,colors='w')

plt.title("Pie Chart")
plt.show()
```
