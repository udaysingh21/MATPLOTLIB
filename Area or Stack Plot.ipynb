# Area or Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area1=[4,2,1,5,3]
area2=[3,2,1,5,4]
area3=[3,3,4,1,5]

labels=['area1','area2','area3']
colors=['r','g','b']
plt.stackplot(x,area1,area2,area3,labels=labels,colors=colors)

plt.title('Python')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# plt.grid()

plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[4,2,1,5,3]
area2=[3,2,1,5,4]
area3=[3,3,4,1,5]

labels=['area1','area2','area3']
colors=['r','g','b']
plt.stackplot(x,area1,area2,area3,labels=labels,colors=colors)

plt.title('Python')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid()

plt.legend(loc=2)
plt.show()
```
