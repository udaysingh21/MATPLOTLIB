# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=['Python','JS','Java','C++','C']
y=[98,95,87,75,60]

plt.xlabel("Programming Languages",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.title("Programming Language Demand",fontsize=15)

colors=['r','g','b','m','y']
bars=plt.bar(x,y,width=0.5,color=colors,align='center',edgecolor='black',linewidth=2,linestyle=':',alpha=0.8)

labels=['Python','JS','Java','C++','C']
plt.legend(bars,labels)


plt.show()
```

**Multiple Bars in single BAR PLOT**

```python
x=['Python','JS','Java','C++','C']
y=[98,95,87,75,60]
z=[45,76,93,81,63]

width=0.4
p=np.arange(len(x)) # Returns evenly spaced values within given intervals
p1=[j+width for j in p]

plt.xlabel("Programming Languages",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.title("Programming Language Demand",fontsize=15)

colors=['r','g','b','m','y']
colors1=['y','c','g','r','m']
bars=plt.bar(x,y,width,color=colors,align='center',edgecolor='black',linewidth=2,linestyle=':',alpha=0.8)
bars1=plt.bar(p1,z,width,color=colors1,align='center',edgecolor='black',linewidth=2,alpha=0.8)

labels=['Python','JS','Java','C++','C']
plt.legend(bars,labels)

plt.xticks(p+width/2,x)
plt.show()
```

**Horizontal Bar**

```python
x=['Python','JS','Java','C++','C']
y=[98,95,87,75,60]
z=[45,76,93,81,63]

width=0.4
p=np.arange(len(x)) # Returns evenly spaced values within given intervals
p1=[j+width for j in p]

plt.xlabel("Programming Languages",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.title("Programming Language Demand",fontsize=15)

colors=['r','g','b','m','y']
colors1=['y','c','g','r','m']
bars=plt.barh(x,y,width,color=colors,align='center',edgecolor='black',linewidth=2,linestyle=':',alpha=0.8)
bars1=plt.barh(p1,z,width,color=colors1,align='center',edgecolor='black',linewidth=2,alpha=0.8)

labels=['Python','JS','Java','C++','C']
plt.legend(bars,labels)

plt.xticks(p+width/2,x)
plt.show()
```
