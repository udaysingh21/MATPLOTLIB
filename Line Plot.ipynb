# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y=np.array([20,50,40,30,50,40,30,70])

plt.plot(y,color='red',linewidth=2,marker='o',mec='black',alpha=0.6)
plt.show()
```

```python
y=np.array([20,50,40,30,50,40,30,70])
z=np.array([10,60,20,40,70,20,50,30])

plt.plot(y,color='r',marker='o')
plt.plot(z,color='c',linestyle="--",marker='d')
plt.legend(['women','men'],loc='lower right')
plt.show()
```
